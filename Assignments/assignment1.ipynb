# Data Normalization and Formatting
 - Improved Model Performance
 - Avoidance of Biased Results
 - Faster Convergence
 - Interpretable Coefficients
 - Compatibility with Certain Algorithms

```python
df.head()
```

```python
# Min max
min_max_scaler = MinMaxScaler()
df['age'] = min_max_scaler.fit_transform(df[['age']])
```

```python
# Min-max normalization for column 'A' without using Sklearn
min_val = df['A'].min()
max_val = df['A'].max()
df['A_min_max'] = (df['A'] - min_val) / (max_val - min_val)
```

```python
df.head()
```

```python
# Scale just one column using z-score normalization
z_score_scaler = StandardScaler()
df['rm'] = z_score_scaler.fit_transform(df[['rm']])
```

```python
# Z-score normalization for column 'A'
mean_val = df['A'].mean()
std_val = df['A'].std()
df['A_z_score'] = (df['A'] - mean_val) / std_val
```
